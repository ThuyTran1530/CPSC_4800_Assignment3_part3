```python
# Scatter plot
def scatterplot(HousePrice, variables, columns = 2):
    # Set dynamic of plot
    num_vars = len(variables)
    rows = math.ceil(num_vars / columns)
    plt.figure(figsize = (columns * 5, rows * 5))
    
    # Loop
    for i, var in enumerate(variables):
        plt.subplot(rows, columns, i +1)
        sns.scatterplot( x = HousePrice[var], y = HousePrice['SalePrice'], hue = HousePrice['SalePrice'], palette = 'crest', legend = None)
        sns.regplot(x = HousePrice[var], y = HousePrice['SalePrice'], scatter = False, color = 'blue', line_kws = {'color': 'red'})
        plt.title(f'Scatter plot of Selling Price & {var}')
        plt.xlabel(var)
        plt.ylabel('Selling Price')
    plt.tight_layout()
    plt.show()
    
variables = top_variables
scatterplot(HousePrice, variables, columns = 2)
```
